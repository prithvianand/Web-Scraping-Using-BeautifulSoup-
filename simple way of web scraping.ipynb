```python
import requests
import bs4
import html5lib
```

```python
url = 'https://www.amazon.in/Test-Exclusive-718/dp/B07DJD1XZ4'
r = requests.get(url)
soup = bs4.BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
title = soup.find(id="productTitle" )
print(title.text)
```

```python
price = soup.find(id="priceblock_dealprice")
print(price.text)
```
